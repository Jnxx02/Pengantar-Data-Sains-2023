```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_excel('/content/drive/Othercomputers/My Laptop/Kuliah/SEMESTER 3/Pengantar Data Sains/Pengantar-Data-Sains-2023/Pertemuan 9/datarumah.xlsx')
data.head()
```

```python
import seaborn as sns
import matplotlib as plt
```

```python
from sklearn import linear_model
```

```python
reg = linear_model.LinearRegression()

harga_rumah = data[['Harga rumah (1000s)(y)']]
luas_rumah = data[['Feet Persegi(x)']]

reg.fit(luas_rumah, harga_rumah)
```

```python
b1 = reg.coef_
b0 = reg.intercept_
print (b0, b1)
```

```python
sns.scatterplot(data, x = 'Feet Persegi(x)', y = 'Harga rumah (1000s)(y)')
```

```python
y_pred = reg.predict(luas_rumah)
y_pred
```

```python
# sns.lineplot(data, x = 'Feet Persegi(x)', y = 'Harga rumah (1000s)(y)')
sns.regplot(data, x = 'Feet Persegi(x)', y = 'Harga rumah (1000s)(y)')
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
mae = mean_absolute_error(harga_rumah, y_pred)
mse = mean_squared_error(harga_rumah, y_pred)
rmse = np.sqrt(mse)
r2 = r2_score(harga_rumah, y_pred)

print('MAE: ', mae)
print('MSE: ', mse)
print('RMSE: ', rmse)
print('R2: ', r2)
```
